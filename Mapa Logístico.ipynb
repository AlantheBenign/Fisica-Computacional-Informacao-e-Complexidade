```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def mapa(x,r):
    res = r*x*(1-x)
    return res
```

```python
limite = 51
n = np.arange(limite)

rs = [0.5,2.5,3.1,3.5,3.7]

fig, ((ax0, ax1), (ax2, ax3), (ax4, ax5)) = plt.subplots(3, 2,figsize=(12, 12))
fig.suptitle('x = 0.5')

for j in range(5):
    x = 0.5
    xs = np.array([])
    xs = np.append(xs,x)
    for i in range(limite-1):
        x = mapa(x,rs[j])
        xs = np.append(xs,x)
        
    if j==0:
        ax0.plot(n,xs,label='r = ' + str(rs[j]))
        ax1.plot(n,xs,label='r = ' + str(rs[j]))
    elif j==1:
        ax2.plot(n,xs,label='r = ' + str(rs[j]))
    elif j==2:
        ax3.plot(n,xs,label='r = ' + str(rs[j]))
    elif j==3:
        ax4.plot(n,xs,label='r = ' + str(rs[j]))
    elif j==4:
        ax5.plot(n,xs,label='r = ' + str(rs[j]))

for ax in fig.get_axes():
    ax.label_outer()
    ax.legend()
```

Aqui percebemos que as popula√ß√µes convergem para um valor na situa√ß√£o de equil√≠brio ou oscilam entre alguns (ou v√°rios) valores conforme as gera√ß√µes se passam.

```python
limite = 51
n = np.arange(limite)

rs = [0.5,2.5,3.1,3.5,3.7]
x0s = [0.5,0.25,0.75]

fig, ((ax0, ax1), (ax2, ax3), (ax4, ax5)) = plt.subplots(3, 2,figsize=(12, 12))
fig.suptitle('x = 0.25, 0.50, 0.75')

for j in range(5):
    plt.figure()
    for k in range(3):
        x = x0s[k]
        xs = np.array([])
        xs = np.append(xs,x)
        for i in range(limite-1):
            x = mapa(x,rs[j])
            xs = np.append(xs,x)
        if j==0:
            ax0.plot(n,xs,label='r = ' + str(rs[j]) + ' , x0 = ' + str(x0s[k]))
            ax1.plot(n,xs,label='r = ' + str(rs[j]) + ' , x0 = ' + str(x0s[k]))
        elif j==1:
            ax2.plot(n,xs,label='r = ' + str(rs[j]) + ' , x0 = ' + str(x0s[k]))
        elif j==2:
            ax3.plot(n,xs,label='r = ' + str(rs[j]) + ' , x0 = ' + str(x0s[k]))
        elif j==3:
            ax4.plot(n,xs,label='r = ' + str(rs[j]) + ' , x0 = ' + str(x0s[k]))
        elif j==4:
            ax5.plot(n,xs,label='r = ' + str(rs[j]) + ' , x0 = ' + str(x0s[k]))

for ax in fig.get_axes():
    ax.label_outer()
    ax.legend(loc=4)
```

Aqui percebemos que para popula√ß√µes iniciais bastante distintas o comportamento dessas popula√ß√µes n√£o √© muito alterado conforme as gera√ß√µes avan√ßam.

```python
limite = 1000
pontos_por_r = 100
divisoes = 2500
poucos = 200

rs = np.linspace(0,3,poucos)
rs = np.append(rs,np.linspace(3,4,divisoes))
xss = np.array([])
rss = np.array([])

for j in rs:
    x = 0.5
    clear_output(wait=True)
    print(str("%.2f" % ((j/4)*100))+"%", end="\n")
    for i in range(limite):
        x = mapa(x,j)
        if i >= (limite - pontos_por_r - 1):
            rss = np.append(rss,j)
            xss = np.append(xss,x)

xss = np.reshape(xss,(divisoes+poucos,pontos_por_r + 1))
rss = np.reshape(rss,(divisoes+poucos,pontos_por_r + 1))
```

```python
plt.figure()
for i in range(divisoes+poucos):
    plt.plot(rss[i],xss[i],'bo',markersize=0.5)
plt.show()
```

O gr√°fico acima mostra os valores que as popula√ß√µes assumem ap√≥s v√°rias gera√ß√µes:

Para valores de r < 1, as popula√ß√µes se extinguem; 

Para 1 < r < 3, as popula√ß√µes convergem para um √∫nico valor est√°vel;

Para 3 < r < ~3.4, as popula√ß√µes ocilam entre dois valores poss√≠veis;

Para a maioria dos demais intervalos as popula√ß√µes oscilam entre in√∫meros valores de forma ca√≥tica;

```python
plt.figure()
for i in range(divisoes+poucos):
    if 3.7 <= rss[i][0] <= 3.9:
        plt.plot(rss[i],xss[i],'bo',markersize=0.5)
plt.show()
```

Aqui percebemos que existem comporamentos diferentes dentro da regi√£o ca√≥tica.

```python
plt.figure()
for i in range(divisoes+poucos):
    if 3.840 <= rss[i][0] <= 3.856:
        plt.plot(rss[i],xss[i],'bo',markersize=0.5)
plt.ylim((0.425,0.575))
plt.show()
```

Para este intervalo de r, o comportamento √© semelhante ai da curva como um todo, mostrando seu comportamento fractal.
